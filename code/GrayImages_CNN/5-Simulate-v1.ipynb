```python
import tensorflow as tf
import scipy.misc
import model
# from subprocess import call
# import time
import csv
import numpy as np
import config as cfg
import cv2

sess = tf.InteractiveSession()
saver = tf.train.Saver()
saver.restore(sess, "save/model.ckpt")

xs = []
ys = []

increased_path = '../dataset/20200512/3-increased'
with open(increased_path + '/data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        # print(row[0], row[1])
        xs.append(increased_path + '/' + row[0])
        ys.append(row[1])

        
increased_path = '../dataset/20200513/3-increased'
with open(increased_path + '/data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        # print(row[0], row[1])
        xs.append(increased_path + '/' + row[0])
        ys.append(row[1])        
"""
increased_path = '../dataset/20200414-revised/3-increased'
with open(increased_path + '/data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        # print(row[0], row[1])
        xs.append(increased_path + '/' + row[0])
        ys.append(row[1])

increased_path = '../dataset/20200420-revised/3-increased'
with open(increased_path + '/data.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        # print(row[0], row[1])
        xs.append(increased_path + '/' + row[0])
        ys.append(row[1])
"""        


# print(xs[0])

"""
# 이미지를 읽고, 크기변환, normalize
cv2.resize(cv2.imread(filepath)[cfg.modelheight:], dsize=(200, 66)) / 255.0)

# 예시로 출력 해 보자
# cv2.imshow("src : " + str(y_out[i]), x_out[i])
# print('filepath : ' + filepath)
# cv2.waitKey(0)

def getDirectionString(dir):
    if dir == 0:  # 0 정지!!!
        return 'Stop'
    elif dir == 1:  # 1 = 좌회전
        return 'Left'
    elif int(row[1]) == 2:  # 2 = 전진
        return 'Forward'
    elif int(row[1]) == 3:  # 3 = 우회전
        return 'Right'
    
    return 'Error'
"""

correct_num = 0
missed_forward = 0
missed_left = 0
missed_right = 0
missed_upleft = 0
missed_upright = 0
missed_etc = 0


# while(cv2.waitKey(10) != ord('q')):
for i in range(len(xs)):
    filepath = xs[i]
    # print('filepath : ' + filepath)
    
    # full_image = scipy.misc.imread(filepath, mode="RGB")
    full_image = cv2.imread(filepath)
    full_image_gray = cv2.cvtColor(full_image, cv2.COLOR_BGR2GRAY)
    
    # image_feed2model = scipy.misc.imresize(full_image[cfg.modelheight:], [66, 200]) / 255.0
    image_feed2model = cv2.resize(full_image_gray[cfg.modelheight:], dsize=(200, 66)) / 255.0
    
    # image_largescale = scipy.misc.imresize(full_image[cfg.modelheight:], [66 * 3, 200 * 3])
    image_largescale = cv2.resize(full_image_gray[cfg.modelheight:], dsize=(200*3, 66*3)) / 255.0
    
    degrees = model.y.eval(feed_dict={model.x: [image_feed2model], model.keep_prob: 1.0})
    # print("Label y: " + ys[i], 'prediced value:', np.argmax(degrees, axis=1))
    #if False:
    #if i % 10 == 0:
    if False:
        # cv2.imshow("Feed", cv2.cvtColor(image_largescale, cv2.COLOR_RGB2BGR))
        cv2.imshow("Feed", image_largescale)
        cv2.waitKey(0)
        # cv2.imshow("Original", cv2.cvtColor(full_image, cv2.COLOR_RGB2BGR))
        cv2.imshow("Original", full_image)
        cv2.waitKey(0)

    if int(ys[i]) == np.argmax(degrees, axis=1):
        correct_num += 1
    else:
        if int(ys[i]) == cfg.LEFT:  # LEFT
            missed_left += 1
        elif int(ys[i]) == cfg.UP:  # FORWARD
            missed_forward += 1
        elif int(ys[i]) == cfg.RIGHT:  # RIGHT
            missed_right += 1
        elif int(ys[i]) == cfg.UP_LEFT:
            missed_upleft += 1
        elif int(ys[i]) == cfg.UP_RIGHT:
            missed_upright += 1
        else:  # etc
            missed_etc = missed_etc + 1

    if (i > 0) and (i % 100 == 0):
        print('확인한 사진 수 :', i, '| 정확도: ', round((correct_num/i)*100, 2))


# summary
print('전체 이미지 수 ' + str(len(xs)))
print('LEFT 를 못맞춘 경우의 수: ' + str(missed_left))
print('RIGHT 를 못맞춘 경우의 수: ' + str(missed_right))
print('FORWARD 를 못맞춘 경우의 수: ' + str(missed_forward))
print('FORWARD-LEFT 를 못맞춘 경우의 수: ' + str(missed_upright))
print('FORWARD-RIGHT 를 못맞춘 경우의 수: ' + str(missed_upleft))
print('ETC 를 못맞춘 경우의 수: ' + str(missed_etc))

cv2.destroyAllWindows()

# THE END

```
